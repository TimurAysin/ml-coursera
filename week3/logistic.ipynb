```python
import pandas as pd
import numpy as np
from math import exp
from sklearn.metrics import roc_auc_score
```

```python
data = pd.read_csv("data-logistic.csv", header=None)
```

```python
X = data.iloc[:, 1:]
y = data.iloc[:, :1]
y
```

```python
X
X.iloc[0, 0]
```

```python
def gradient_step(X, y, w1, w2, C = 0):
    k = 0.1
    l = X.shape[0]
    
    sum1 = 0
    sum2 = 0

    for i in range(l):
        yi = y.iloc[i, 0]
        xi = X.iloc[i, 0]
        sum1 += yi * xi * (1 - 1 / (1 + exp(-yi * (w1 * X.iloc[i, 0] + w2 * X.iloc[i, 1]))))

        xi = X.iloc[i, 1]
        sum2 += yi * xi * (1 - 1 / (1 + exp(-yi * (w1 * X.iloc[i, 0] + w2 * X.iloc[i, 1]))))
    w1 = w1 + k / l * sum1 - k * C * w1
    w2 = w2 + k / l * sum2 - k * C * w2

    return w1, w2
```

```python
def gradient_descend(X, y, C=0):
    w1 = 0
    w2 = 0
    maxiter = 10000
    eps = 1e-5

    for i in range(maxiter):
        nw1, nw2 = gradient_step(X, y, w1, w2, C)
        if ((w1 - nw1) ** 2 + (w2 - nw2) ** 2) ** (1 / 2) < eps:
            return w1, w2
        else:
            w1, w2 = nw1, nw2
    return w1, w2
```

```python
w1, w2 = gradient_descend(X, y)
w1, w2
```

```python
y_scores = []

for i in range(X.shape[0]):
    y_scores.append(1 / (1 + exp(-w1 * X.iloc[i, 0] - w2 * X.iloc[i, 1])))
```

```python
y_scores
```

```python
roc_auc_score(y, y_scores)
```

```python
roc_auc_score(y, y_scores)
```
