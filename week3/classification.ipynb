```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve
```

```python
data = pd.read_csv("classification.csv")
data
```

```python
tp = 0
fp = 0
fn = 0
tn = 0

y_true = data.iloc[:, 0]
y_pred = data.iloc[:, 1]

for i in range(y_true.shape[0]):
    value_true = y_true[i]
    value_pred = y_pred[i]
    
    if value_true == 1 and value_pred == 1:
        tp += 1
    elif value_true == 0 and value_pred == 0:
        tn += 1
    elif value_true == 1 and value_pred == 0:
        fn += 1
    else:
        fp += 1

print(tp, fp, fn, tn)
        
with open("classification", "w") as f:
    f.write("43 34 59 64")
```

```python
print(accuracy_score(y_true, y_pred))
print(precision_score(y_true, y_pred))
print(recall_score(y_true, y_pred))
print(f1_score(y_true, y_pred))

with open("scores", "w") as f:
    f.write("0.54 0.56 0.42 0.48")
```

```python
data = pd.read_csv("scores.csv")
data
```

```python
y_true = data["true"]
score_logreg = data["score_logreg"]
score_cvm = data["score_svm"]
score_knn = data["score_knn"]
score_tree = data["score_tree"]

print(roc_auc_score(y_true, score_logreg))
print(roc_auc_score(y_true, score_cvm))
print(roc_auc_score(y_true, score_knn))
print(roc_auc_score(y_true, score_tree))

with open("column", "w") as f:
    f.write("score_logreg")
```

```python
scores = [score_logreg, score_cvm, score_knn, score_tree]

for score in scores:
    precision, recall, thresholds = precision_recall_curve(y_true, score)    
    df = pd.DataFrame({"precision": precision, "recall": recall})
    print(df[df['recall'] >= 0.7].max()['precision'])
    
with open("precision", "w") as f:
    f.write("score_tree")
```
